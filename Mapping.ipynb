```python
import folium
import os
import json
import pandas as pd
```

```python
ny_zip_codes = os.path.join(r'C:\Users\Evan\Desktop\Capstone 2\Data\work_plz.geojson')
with open(ny_zip_codes) as zip_codes:
    zip_data = json.load(zip_codes)


italian_data = pd.read_csv(r'C:\Users\Evan\Desktop\Capstone 2\Data\italian_complete.csv', index_col='Unnamed: 0')
```

```python
italian_data['postalcode'] = italian_data['postalcode'].astype('str')
```

```python
m = folium.Map(
        location=[40.73942, -73.98160],
        zoom_start=10
)
```

```python
m.choropleth(
    geo_data=zip_data,
    name='choropleth',
    data=italian_data,
    columns=['postalcode', 'italian_prop'],
    key_on='feature.properties.postalcode',
    fill_color='YlGn',
    fill_opacity=1,
    line_opacity=1,
    legend_name='Test'
)
```

```python
folium.LayerControl().add_to(m)
```
